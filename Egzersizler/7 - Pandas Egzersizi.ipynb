# The Dataset From This [Address-1](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv) & [Adres-2](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv)

```python
import numpy as np
import pandas as pd
```

```python
url1 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv"
url2 = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv"

cars1 = pd.read_csv(url1)
cars1 = cars1.dropna(axis = 1)
cars2 = pd.read_csv(url2)
```

```python
cars1.head()
```

```python
cars2.head()
```

```python
print(cars1.shape)
print(cars2.shape)
```

```python
cars = pd.concat([cars1, cars2])
cars.head()
```
