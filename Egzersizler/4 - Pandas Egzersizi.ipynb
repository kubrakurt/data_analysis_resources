# The Dataset From This [Address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv)

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"

df = pd.read_csv(url)
df.head()
```

## Ortalama olarak hangi kıta daha fazla bira tüketiyor?

```python
df.groupby("continent").mean().sort_values(by = "beer_servings", ascending = False)["beer_servings"][0]
```

```python
df.groupby(by = "continent")["beer_servings"].mean().sort_values(ascending = False)[0]
```

## Şarap tüketiminin kıtalara göre istatistikleri nedir?

```python
df.groupby("continent")["wine_servings"].describe()
```

## Her kıtanın her sütunu için medyan nedir?

```python
df.groupby("continent").median().T
```

## Spirit_Servings değişkeni için kıtalara göre mean, max ve min nasıl bakarız?

```python
df.groupby("continent")["spirit_servings"].agg(["mean","min","max"]).T
```
