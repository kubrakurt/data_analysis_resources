# The Dataset From This [Address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user)

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

df = pd.read_csv(url, sep = "|")
df.head()
```

## Meslek başına ortalama yaş kaçtır?

```python
df.groupby("occupation")[["age"]].mean()
```

## Meslek başına erkeklerin oranı kaçtır?

```python
male = df[df["gender"] == "M"].groupby("occupation").count()
female = df[df["gender"] == "F"].groupby("occupation").count()
(male["gender"] / (male["gender"] + female["gender"])).sort_values(ascending = False)
```

## Her meslek için min ve max yaşlar kaçtır?

```python
df.groupby("occupation")["age"].agg(["min","max"])
```

## Meslekler ve cinsiyetler için ortalama yaşlar kaçtır?

```python
df.groupby(["occupation","gender"])[["age"]].mean()
```
