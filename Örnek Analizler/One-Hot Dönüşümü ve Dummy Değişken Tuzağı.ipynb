# One-Hot Dönüşümü ve Dummy Değişken Tuzağı

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df_one_hot = pd.get_dummies(df, columns = ["sex"], prefix = ["sex"])
df_one_hot.head()
```
